# Multi Layer Perceptron (MLP)

```python
import matplotlib.pyplot as plt
import numpy as np


def relu(x):
    return np.maximum(0, x)

def perceptron(w: np.ndarray, x: np.ndarray):
    return relu(np.dot(w, x))

def donut_datasett(n_ring=300, n_center=150, r_inner=2.0, r_outer=3.2, noise=0.15, seed=0):
    rng = np.random.default_rng(seed)
    r = rng.uniform(r_inner, r_outer, n_ring) + rng.normal(0, noise, n_ring)
    theta = rng.uniform(0, 2*np.pi, n_ring)
    X_ring = np.c_[r*np.cos(theta), r*np.sin(theta)]
    y_ring = np.ones(n_ring, dtype=int)

    X_center = rng.normal(0, r_inner*0.35, size=(n_center, 2))
    X_center += rng.normal(0, noise, size=X_center.shape)
    y_center = np.zeros(n_center, dtype=int)

    X = np.vstack([X_center, X_ring])
    y = np.hstack([y_center, y_ring])
    return X, y

def tegn_donut():
    X, y = donut_datasett()

    plt.figure(figsize=(6,6))
    plt.scatter(X[y==0,0], X[y==0,1], s=25, label="Senter (0)")
    plt.scatter(X[y==1,0], X[y==1,1], s=25, label="Donut (1)")
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)
    plt.xlabel("x₁"); plt.ylabel("x₂")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.axis('equal')
    plt.tight_layout()
    plt.show()


tegn_donut()
```
